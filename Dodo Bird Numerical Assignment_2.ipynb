## Dodo Bird Population Differential Equations

In this notebook we will solve a simple population equation [1]

```python
## Library
import numpy as np
import matplotlib.pyplot as plt
```

## Dodo Bird Population Equation
The Dodo bird population differential equation is given by:
\begin{equation}
\frac{d y}{dt}= y(t)*(1 - x(t))
\end{equation}
for $0\leq t \leq 10$ years, where $x$ is the subsequent Human-Dodo bird interactions and $y$ is the population of Dodo, with the initial condition
$$y(0)=1000.$$

The Human Population differential equation is given by:
\begin{equation}
\frac{d x}{dt}= 0.1*(x(t))
\end{equation}
for $0\leq t \leq 10$ years, where $x$ is the subsequent interactions with the initial condition
$$x(0)=1.$$



Note : 

The exact solution to Human Population differential equation is given by :
\begin{equation}
x(t) = e^{0.1*t}
\end{equation}

The exact solution to Dodo Bird population differential equation is given by :
\begin{equation}
y(t) = e^{t-10e^{0.1t} + c}
\end{equation}

where c = 16.9077553

As the equation is non-linear for the first case, we must numerically approximate the solution.
Thus in this simulation we set the difference equations with the inital condition $y(0)=1000$, over the time period $0\leq t \leq 10$ years.

## Discrete Domain 
To numerically approximate the solution we discretise the continuous domain using a step size, $h=0.1$, 
which gives
\begin{equation}
t_i=0+i h,
\end{equation}
for $i=0,...100$.
The figure below illustrate the discrete domain.

```python
h=0.1
t_end=10
t=np.arange(0,t_end+h/2,h)
fig = plt.figure(1,figsize=(5,4))
ax = fig.add_subplot(1,1,1)
ax.plot(t,0*t,'o:',color='k')
ax.set_xlabel('Time (yrs)')
plt.title('Discrete Time Domain')
plt.show()
```

```python
N=int(t_end/h) # Number of Steps
Y=np.zeros(N+1)
X=np.zeros(N+1)
Y[0]=1000
X[0]=1
```

We numerically approximate the Dodo bird population differential equation using Eulers method to give the difference equation
\begin{equation}
\frac{Y_{i+1}-Y_i}{h}=Y_{i}(1 - X_{i})
\end{equation}
re-arranging
\begin{equation}
\color{red}{Y_{i+1}}=Y_i+h\big(Y_{i}(1 - X_{i})\big)
\end{equation}

where $h$ is the stepsize and where $Y_i$ is the numerical approximation of $y(t_i)$ at time $t_i$, for $i=0,...,10$

Similarly the numerical approximation of Human population differential equation using Eulers method gives the difference equation

\begin{equation}
\frac{X_{i+1}-X_i}{h}=0.1X_{i}
\end{equation}
re-arranging
\begin{equation}
\color{red}{X_{i+1}}=X_i+h\big(0.1X_{i}\big)
\end{equation}

where $h$ is the stepsize and where $X_i$ is the numerical approximation of $x(t_i)$ at time $t_i$, for $i=0,...,10$


```python
for i in range (0,N):
    Y[i+1]=Y[i]+h*(Y[i]*(1-X[i]))# Numerical Solution Y
    X[i+1]=X[i]+h*(0.1*X[i])# Numerical Solution X

y = np.exp(t - 10*np.exp(0.1*t) + 16.9077553) # Analytical Solution Y 
x = np.exp(0.1*t)
```

## Results : Dodo Bird  Population
The plot below shows the Numerical solution $Y_i$ of the Dodo Bird Logistic Population as a function of time $t$.

```python
## Plotting Figure
fig = plt.figure(1,figsize=(12,4))

# --- left hand plot
ax = fig.add_subplot(1,3,1)
ax.plot(t,Y,'o:',color='blue')
ax.plot(t,0*t,'--k')


ax.set_xlabel('Time (yrs)')
ax.set_ylabel('Population (billions)')
#ax.legend(loc='best')
ax.set_title('Numerical Solution')

# --- middle plot
ax = fig.add_subplot(1,3,2)
ax.plot(t,y,'-',color='green')
ax.plot(t,0*t,'--k')

ax.set_xlabel('Time (yrs)')
ax.set_ylabel('Population (billions)')
ax.set_title('Analytic (exact) Solution')

# --- right hand plot
ax = fig.add_subplot(1,3,3)
ax.plot(t,y-Y,'o:',color='red')
ax.set_title('Error')
ax.set_ylabel('Error(billions)')
ax.set_xlabel('Time (yrs)')

# --- title, explanatory text and save
fig.suptitle('Leaky Solution for Dodo Bird Population Differential Equation', fontsize=20)
plt.tight_layout()
plt.subplots_adjust(top=0.85)
```

## Results : Human - Population

The plot below shows the Numerical solution $X_i$ of the Human Population as a function of time $t$.

```python
## Plotting Figure
fig = plt.figure(1,figsize=(12,4))

# --- left hand plot
ax = fig.add_subplot(1,3,1)
ax.plot(t,X,'o:',color='blue')
ax.plot(t,0*t,'--k')


ax.set_xlabel('Time (yrs)')
ax.set_ylabel('Population (billions)')
#ax.legend(loc='best')
ax.set_title('Numerical Solution')

# --- middle plot
ax = fig.add_subplot(1,3,2)
ax.plot(t,x,'-',color='green')
ax.plot(t,0*t,'--k')

ax.set_xlabel('Time (yrs)')
ax.set_ylabel('Population (billions)')
ax.set_title('Analytic (exact) Solution')

# --- right hand plot
ax = fig.add_subplot(1,3,3)
ax.plot(t,x-X,'o:',color='red')
ax.set_title('Error')
ax.set_ylabel('Error(billions)')
ax.set_xlabel('Time (yrs)')

# --- title, explanatory text and save
fig.suptitle('Leaky Solution for Human - Dodo Bird Interaction Differential Equation', fontsize=20)
plt.tight_layout()
plt.subplots_adjust(top=0.85)
```
